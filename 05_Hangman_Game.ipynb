**Project 05: Hangman Game - Python**

```python
import random

# Hangman stages graphics
HANGMAN_PICS = [
    """
       ------
       |    |
       |
       |
       |
       |
    =========
    """,
    """
       ------
       |    |
       |    O
       |
       |
       |
    =========
    """,
    """
       ------
       |    |
       |    O
       |    |
       |
       |
    =========
    """,
    """
       ------
       |    |
       |    O
       |   /|
       |
       |
    =========
    """,
    """
       ------
       |    |
       |    O
       |   /|\\
       |
       |
    =========
    """,
    """
       ------
       |    |
       |    O
       |   /|\\
       |   /
       |
    =========
    """,
    """
       ------
       |    |
       |    O
       |   /|\\
       |   / \\
       |
    =========
    GAME OVER!
    """
]

# Word bank for the game
WORD_LIST = ["python", "hangman", "developer", "computer", "artificial", "intelligence", "challenge", "javascript", "science", "machine", "learning"]

def get_random_word():
    """Selects a random word from the word list"""
    return random.choice(WORD_LIST).upper()

def display_word_progress(word, guessed_letters):
    """Displays the word with guessed letters and underscores for unknown ones"""
    return " ".join([letter if letter in guessed_letters else "_" for letter in word])

def hangman():
    print("\n🎭 Welcome to Hangman! 🔤")
    print("🎯 Try to guess the secret word before the Hangman is complete! ⏳\n")

    word = get_random_word()
    guessed_letters = set()
    attempts = 6

    while attempts > 0:
        print(HANGMAN_PICS[6 - attempts])
        print("\n🔡 Word:", display_word_progress(word, guessed_letters))
        print(f"💡 Letters Guessed: {', '.join(sorted(guessed_letters)) if guessed_letters else 'None'}")
        print(f"❤️ Attempts Left: {attempts}\n")

        guess = input("✏️ Enter a letter: ").upper()

        if len(guess) != 1 or not guess.isalpha():
            print("⚠️ Please enter a single valid letter!\n")
            continue

        if guess in guessed_letters:
            print("⚠️ You already guessed this letter! Try another one.\n")
            continue

        guessed_letters.add(guess)

        if guess in word:
            print("✅ Correct guess! 🎉\n")
            if set(word) <= guessed_letters:
                print(f"🏆 CONGRATULATIONS! You guessed the word: {word} 🎉")
                break
        else:
            print("❌ Incorrect guess! 😢\n")
            attempts -= 1

    if attempts == 0:
        print(HANGMAN_PICS[-1])
        print(f"💀 You lost! The correct word was: {word} \n")

    # Ask if user wants to play again
    if input("🔄 Do you want to play again? (yes/no): ").lower() == "yes":
        hangman()
    else:
        print("\n👋 Thanks for playing Hangman! See you next time!")

# Start the game
hangman()
```
