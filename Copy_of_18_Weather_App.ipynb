```python
# Install necessary libraries
!pip install requests
!pip install matplotlib

import requests
import datetime
import matplotlib.pyplot as plt
from ipywidgets import interact

# Step 1: Set up API key and endpoint
API_KEY = 'b1a3e85995a21e9c234dd230738d24f7'  # Replace with your OpenWeatherMap API key
BASE_URL = 'http://api.openweathermap.org/data/2.5/'

# Step 2: Function to get weather data for a city
def get_weather_data(city):
    # URL to fetch weather data
    url = f"{BASE_URL}weather?q={city}&appid={API_KEY}&units=metric"
    response = requests.get(url)

    if response.status_code == 200:
        return response.json()
    else:
        return None

# Step 3: Function to get forecast data (next 3-5 days)
def get_forecast_data(city):
    # URL to fetch forecast data (5 days)
    url = f"{BASE_URL}forecast?q={city}&appid={API_KEY}&units=metric"
    response = requests.get(url)

    if response.status_code == 200:
        return response.json()
    else:
        return None

# Step 4: Function to display weather information
def display_weather_info(city):
    weather_data = get_weather_data(city)
    if weather_data:
        print(f"Weather Information for {city.capitalize()}:")
        print(f"Temperature: {weather_data['main']['temp']}°C")
        print(f"Weather: {weather_data['weather'][0]['description']}")
        print(f"Humidity: {weather_data['main']['humidity']}%")
        print(f"Wind Speed: {weather_data['wind']['speed']} m/s")
        print(f"Pressure: {weather_data['main']['pressure']} hPa")
    else:
        print(f"City {city} not found!")

# Step 5: Function to display forecast (next 3-5 days)
def display_forecast_info(city):
    forecast_data = get_forecast_data(city)
    if forecast_data:
        print(f"Forecast for {city.capitalize()}:")
        for forecast in forecast_data['list'][:5]:  # First 5 forecasts (for 3-5 days)
            dt = datetime.datetime.fromtimestamp(forecast['dt'])
            temp = forecast['main']['temp']
            weather = forecast['weather'][0]['description']
            print(f"{dt.strftime('%Y-%m-%d %H:%M:%S')} - Temp: {temp}°C, Weather: {weather}")
    else:
        print(f"City {city} not found!")

# Step 6: Plotting weather forecast
def plot_weather_forecast(city):
    forecast_data = get_forecast_data(city)
    if forecast_data:
        times = [datetime.datetime.fromtimestamp(f['dt']).strftime('%Y-%m-%d %H:%M:%S') for f in forecast_data['list'][:5]]
        temps = [f['main']['temp'] for f in forecast_data['list'][:5]]

        plt.figure(figsize=(10, 6))
        plt.plot(times, temps, marker='o', color='b')
        plt.title(f'Weather Forecast for {city.capitalize()}')
        plt.xlabel('Date and Time')
        plt.ylabel('Temperature (°C)')
        plt.xticks(rotation=45)
        plt.grid(True)
        plt.tight_layout()
        plt.show()
    else:
        print(f"City {city} not found!")

# Step 7: Create interactive widget for city input
def get_city_from_user(city):
    display_weather_info(city)
    display_forecast_info(city)
    plot_weather_forecast(city)

# Step 8: Use interact to let user choose a city
cities_list = ['London', 'New York', 'Tokyo', 'Paris', 'Berlin', 'Sydney','Karachi','Hyderabad','Lahore', 'Islamabad', 'Rawalpindi', 'Peshawar', 'Quetta', 'Multan', 'Faisalabad', 'Sialkot', 'Gujranwala', 'Abbottabad', 'Sukkur', 'Bahawalpur', 'Dera Ghazi Khan', 'Larkana', 'Muzaffarabad', 'Swat', 'Mardan', 'Chiniot', 'Jhelum'] # Add more cities as needed
interact(get_city_from_user, city=cities_list)
```
